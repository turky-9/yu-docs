# プログラムで微分してみる

とりあえず以下の関数を考える  
$
y = x^2 + 2x + 2
$

グラフは以下となる

```python
import matplotlib.pyplot as plt
import numpy as np
# if using a jupyter notebook
%matplotlib inline

# -4から4までのxを100個用意する
x = np.linspace(-4, 4, 100)

# これが対象の関数
y = x**2 + 2*x + 2

# これ以降はグラフの描画処理
plt.plot(x, y)
plt.grid()
plt.axhline(0, linewidth=2, color="gray")
plt.axvline(0, linewidth=2, color="gray")
plt.show()
```

この関数で$x = 2$の時の傾きを求めたい。

微分すると下記となる  
$
y' = 2x + 2
$  

$x = 2$を代入すると$y' = 6$となり、接線を含めてグラフ化する

```python
# 0から4までのxを50個用意する
x1 = np.linspace(0, 4, 50)

# 接線の関数
y1 = 6 * x1 - 2

# これ以降はグラフの描画処理
plt.plot(x, y)
plt.plot(x1, y1)
plt.grid()
plt.axhline(0, linewidth=2, color="gray")
plt.axvline(0, linewidth=2, color="gray")
plt.show()
```

ここまでは数学のお話し  

プログラムで微分して接線の傾きを計算してみる  
色々なやり方があると思うがここでは$x$と$x + \varDelta$間の傾きを計算して近似値を求める  

$\varDelta = 0.01$の場合

```python
a = 2
d = 0.01
b1 = (a)**2 + 2*(a) + 2
b2 = (a+d)**2 + 2*(a+d) + 2

(b2 - b1) / d
```

$\varDelta = 0.0001$の場合

```python
a = 2
d = 0.0001
b1 = (a)**2 + 2*(a) + 2
b2 = (a+d)**2 + 2*(a+d) + 2

(b2 - b1) / d
```

当然だが$\varDelta$が小さい方が良い値となっている  

今度は$x - \varDelta/2$と$x + \varDelta/2$間の傾きを計算してみる

$\varDelta = 0.0001$の場合

```python
a = 2
d = 0.0001
hd = d/2
b1 = (a-hd)**2 + 2*(a-hd) + 2
b2 = (a+hd)**2 + 2*(a+hd) + 2

(b2 - b1) / d
```

今回は最後の計算方法が一番良い値を得られた
