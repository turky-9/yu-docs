```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 500)

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.show()
```

```python
y = np.linspace(-10, 10, 500)

plt.plot(y,np.square(y)+(2 * y))
plt.plot(y,np.square(y))
plt.plot(y,(2 * y))

plt.grid()
plt.show()
```

```python
from graphviz import Digraph

dot = Digraph()

# ノードの属性を設定
dot.attr('node', shape='circle')
# エッジの属性を登録
dot.attr('edge', arrowsize='1')
dot.attr('graph', rankdir="LR")
dot.attr(ranksep='1')
dot.attr(nodesep='1.5')
#dot.attr(labelfloat='true')

with dot.subgraph(name='cluster_1') as c:
    c.attr(label='Input')
    c.attr(color='transparent')  
    c.node('inp1', 'x1')
    c.node('inp2', 'x2')
with dot.subgraph(name='cluster_2') as c:
    c.attr(label='Hidden')
    c.attr(color='transparent')  
    c.node('hidden1', 'n1')
    c.node('hidden2', 'n2')
with dot.subgraph(name='cluster_3') as c:
    c.attr(label='Output')
    c.attr(color='transparent')  
    c.node('out1', 'o')

dot.edge('inp1', 'hidden1', headlabel='w11', fontcolor='red', labeldistance='2')
dot.edge('inp1', 'hidden2', headlabel='w21', fontcolor='red', labeldistance='2.5')
dot.edge('inp2', 'hidden1', headlabel='w12', fontcolor='red', labeldistance='2.5')
dot.edge('inp2', 'hidden2', headlabel='w22', fontcolor='red', labeldistance='2')
dot.edge('hidden1', 'out1', headlabel='w31', fontcolor='red', labeldistance='2')
dot.edge('hidden2', 'out1', headlabel='w32', fontcolor='red', labeldistance='2')

# 生成されたソースコードを確認
# print(dot)

# 出力フォーマットを指定
#dot.engine = 'circo'

# レンダリング
dot
```
