# 指数と対数
とりあえず指数も対数も$a != 0, a > 0$とするよ
## 指数
### 指数法則
まぁ、これは指数を考えれば明らかだよね

①$a^m a^n = a^{m+n}$  
②$\dfrac{a^m}{a^n} = a^{m-n}$  
③$(a^m)^n = a^{mn}$  
④$(ab)^n = a^n b^n$

### グラフ

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100)

plt.plot(x, 2**x, label="$2^x$")
plt.plot(x, 3**x, label="$3^x$")

plt.grid()
plt.axhline(0, linewidth=2, color="gray")
plt.axvline(0, linewidth=2, color="gray")
plt.legend()
plt.show()
```

## 対数
### 対数法則
これはちょっと証明しないとね

①$\log_a m + \log_a n = \log_a mn$  
②$\log_a m^p = p \log_a m$  
③$\log_a \dfrac{1}{m} = -\log_a m$  
④$\log_a m - \log_a n = \log_a \dfrac{m}{n}$  
⑤$\log_a 0 = 1$  
⑥$\log_a b = \dfrac{\log_c b}{\log_c a}$
